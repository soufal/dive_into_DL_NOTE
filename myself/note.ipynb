### NOTE

#### 什么是机器学习

不需要知道怎么编写程序去识别某一个东西，只需要收集大量关于这个东西的标记数据，然后定义一些参数，利用机器学习算法从这些标记数据中学习到**“最优参数”**，最后利用这些参数可以识别该东西。

* 什么是参数？
    * 可以调整程序的行为。
* 什么是模型？
    * 任一调整后的参数的程序。

> 学习是一个模型训练的过程，学习什么？学习最优的参数。
> 用数据编程（programming with data）：通过用数据集来确定程序行为的方式。

#### 机器学习中的关键组件

* 数据
* 模型
* 目标函数
* 算法

##### 用于学习的**数据**

* 每个数据集由一个个样本组成，每一个样本由一组特征组成。
* 数据的维数：每个样本的特征类别数量都相同，该特征向量的长度。

> 同机器学习相比，深度学习可以处理不同长度的数据。

> 数据不具有充分的代表性，甚至包含了一些社会偏见时，模型就会可能有偏见。

##### 如何转换数据的**模型**

大多数机器学习会涉及到数据的转换，但深度学习会进行层层数据转换。

##### 一个**目标函数**用来量化模型的有效性

* 又称为“损失函数”。
* 定义模型的优劣程度的度量。
* 预测数值：平方误差。预测值与实际值之差的平方。容易优化。
* 分类问题：最小化错误率。预测与实际情况不符的样本比例。不容易优化。通常优化替代目标。

* 数据集：
    * 训练数据集：用于拟合模型参数。
    * 测试数据集：用于评估拟合的模型。

* 过拟合：训练集上的性能好于测试集。

##### 调整模型参数以优化目标函数的**算法**

用于搜索最佳参数，以最小化损失函数。

> 大多数流行的优化算法基于--**梯度下降**。在每个步骤中，检查每个参数，看如果仅对该参数进行少量变动，训练集损失会朝着哪个方向移动。然后就在减少损失的方向上优化参数。



#### 各种机器学习问题

##### 监督学习（supervised learning）

擅长在“给定输入特征”的情况下预测标签，每一个“特征-标签”都对应一个样本（example）。

在训练参数时，为模型提供了一个数据集，其中每一个样本都有真实的标签。

![监督学习](../img/supervised-learning.svg)
:label:`fig_supervised_learning`

###### 回归（regression）

* 标签取任意数值（离散型）。

> 常见的损失函数为**平方误差损失函数**。

###### 分类（classification）

模型能够预测样本属于哪一个类别（category，class）。

> 常见的损失函数为**交叉熵（cross-entropy）**。将“预期风险”作为损失函数，需要将结果的概率乘以与之相关的收益（或伤害）。
